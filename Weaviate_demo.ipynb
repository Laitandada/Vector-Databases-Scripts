## Embedding Conversion

```python
!pip install azure.ai.inference
```

```python
import os
from typing import List
from azure.ai.inference import EmbeddingsClient
from azure.core.credentials import AzureKeyCredential

class AzureEmbeddingsWrapper:
    def __init__(self):
        self.endpoint = "https://models.github.ai/inference"
        self.model_name = "openai/text-embedding-3-large"
        self.token = GITHUB_TOKEN

        self.client = EmbeddingsClient(
            endpoint=self.endpoint,
            credential=AzureKeyCredential(self.token)
        )

    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        response = self.client.embed(
            input=texts,
            model=self.model_name
        )
        return [item.embedding for item in response.data]

    def embed_query(self, text: str) -> List[float]:
        response = self.client.embed(
            input=[text],
            model=self.model_name
        )
        return response.data[0].embedding
```

```python
embedding = AzureEmbeddingsWrapper()
```

```python
result = embedding.embed_query("Hello")
```

## Vector Database Storage

```python
## connect to weaviate-client

import weaviate
from weaviate.auth import AuthApiKey
from weaviate.classes.config import Configure
import os

weaviate_url = WEAVIATE_CLUSTER
weaviate_api_key = WEAVIATE_API_KEY

with weaviate.connect_to_weaviate_cloud(
    cluster_url=weaviate_url,
    auth_credentials=AuthApiKey(weaviate_api_key),
) as client:

    print("Connected:", client.is_ready())
```

```python
client.connect()
```

```python
## cretate a collection

from weaviate.classes.config import Configure, Property, DataType

collection = client.collections.create(
    name="Chatbot",
    properties=[
        Property(
            name="content",
            data_type=DataType.TEXT
        )
    ],
    vectorizer_config=Configure.Vectorizer.none()  # IMPORTANT
)
```

```python
texts = [chunk.page_content for chunk in text_chunks]

vectors = embedding.embed_documents(texts)

collection = client.collections.get("Chatbot")

with collection.batch.dynamic() as batch:
    for chunk, vector in zip(text_chunks, vectors):
        batch.add_object(
            properties={"content": chunk.page_content},
            vector=vector
        )
```

```python
query = "whats a believers authority"
query_vector = embedding.embed_query(query)
```

```python
collection = client.collections.get("Chatbot")

results = collection.query.near_vector(
    near_vector=query_vector,
    limit=8
)
```

```python
for obj in results.objects:
    print(obj.properties["content"])
```

```python
from langchain_openai import ChatOpenAI
import os

llm = ChatOpenAI(
    base_url="https://models.github.ai/inference",
    api_key=GITHUB_TOKEN,
    model="gpt-4o-mini",  # or any available model
    temperature=0
)
```

```python
llm
```

```python
# 3. build context
context = "\n".join(
    obj.properties["content"] for obj in results.objects
)

```

```python
response = llm.invoke(f"""
You are a helpful assistant.

Use the context to answer the question.

If the answer is partially implied, extract and summarize it.

If the context is completely unrelated, say "I don't know".

Context:
{context}

Question:
{query}

Answer:
""")

print(response)
```
